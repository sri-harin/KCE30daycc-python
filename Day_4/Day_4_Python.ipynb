```python
#CHARACTER TO ASCII
ch = input("Enter a character: ")
print("ASCII value:", ord(ch))
```

```python
#ASCII TO CHARACTER
num = int(input("Enter ASCII value (0–127): "))
print("Character:", chr(num))
```

```python
#NEXT CHARACTER
ch = input("Enter a character: ")
print("Next character:", chr(ord(ch) + 1))
```

```python
#CASE DIFFERENT
print("ASCII of A:", ord('A'))
print("ASCII of a:", ord('a'))
print("Difference:", ord('a') - ord('A'))
```

```python
#MANUAL CONVERSION (UPPER TO LOWER)
ch = input("Enter uppercase letter: ")
print("Lowercase:", chr(ord(ch) + 32))
```

```python
#DIGIT CHECK
ch = input("Enter a character: ")
ascii_val = ord(ch)

print("ASCII value:", ascii_val)

if 48 <= ascii_val <= 57:
    print("It is a digit")
else:
    print("Not a digit")
```

```python
#HIDDEN NUMBERS
ch = input("Enter a digit character: ")
num = ord(ch) - ord('0')
print("Integer value:", num)
```

```python
#SYMBOL FINDER
print("33 ->", chr(33))
print("64 ->", chr(64))
print("35 ->", chr(35))
```

```python
#SIMPLE CIPHER
word = input("Enter 3-letter word: ")
encrypted = ""

for ch in word:
    encrypted += chr(ord(ch) + 1)

print("Encrypted word:", encrypted)
```

```python
#ASCII TABLE SNIPPET
for ch in ['X', 'Y', 'Z']:
    print(ch, "->", ord(ch))
```

```python
#BASIC INCREMENT
n = int(input("Enter a number: "))
n += 1
print("After increment:", n)
```

```python
#BASIC DECREMENT
n = int(input("Enter a number: "))
n -= 1
print("After increment:", n)
```

```python
#POSTFIX OBSERVATIONS
# Python equivalent
x = 5
y = x
x += 1
print("Postfix observation -> x:", x, "y:", y)
```

```python
#PREFIX OBSERVATIONS
# Python equivalent
x = 5
x += 1
y = x
print("Prefix observation -> x:", x, "y:", y)
```

```python
 #EXPRESSION TRIAL FOR INCREMENT
a = 10
b = a
a += 1
a += 1
b = b + a
print("Expression Trial 1 -> a:", a, "b:", b)
```

```python
#EXPRESSION TRIAL FOR DECREMENT
a = 5
a -= 1
b = a - a
print("Expression Trial 2 -> a:", a, "b:", b)
```

```python
#LOOP PREVIEW
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
#CHARACTER INCREMENT
ch = 'S'
ch = chr(ord(ch) + 1)
print("Character after increment:", ch)
```

```python
#MULTIPLE INCREMENTS
n = int(input("Enter a number: "))
n += 1     # n++
n += 1     # ++n
n = n + 1
print("Final value after multiple increments:", n)
```

```python
#DECREMENT CHALLENGE
x = int(input("Enter a number: "))
x -= 1
x -= 1
print("After decreasing by 2:", x)
```
