```python
# Numbers 1 to 10
for i in range(1, 11):
    print(i)
```

```python
# Numbers 10 to 1 
for i in range(10, 0, -1):
    print(i)
```

```python
#  Even Numbers between 1 and 20
for i in range(2, 21, 2): 
    print(i)
```

```python
#  Odd Numbers between 1 and 20
for i in range(1, 21, 2):
    print(i)
```

```python
#  Multiplication Table of N 
N = int(input("Enter a number for multiplication table: "))
for i in range(1, 11):
    print(f"{N} x {i} = {N * i}")
```

```python
#  Sum of N
N = int(input("Enter a number to find sum up to N: "))
total = 0 
for i in range(1, N + 1):
    total += i 
    print("Sum =", total)
```

```python
# Custom Range
start = int(input("Enter start number: "))
end = int(input("Enter end number: "))
for i in range(start, end + 1): 
    print(i)
```

```python
# Step Counter 
start = int(input("Enter start number: "))
end = int(input("Enter end number: "))
step = int(input("Enter step value: "))
for i in range(start, end + 1, step):
    print(i)
```

```python
# The Square Table
for i in range(1, 11):
    print(f"{i}^2 = {i**2}")
```

```python
# Alphabet Loop 
for ch in range(ord('A'), ord('Z') + 1):
    print(chr(ch))
```

```python
# While Counting 
i = 1
while i <= 10:
    print(i) 
    i += 1
```

```python
#  User-Controlled Stop
num = int(input("Enter a number (0 to stop): ")) 
while num != 0:
    print("You entered:", num)
    num = int(input("Enter a number (0 to stop): "))
```

```python
#  Sum of Input
total = 0 
num = int(input("Enter a number (negative to stop): "))
while num >= 0:
    total += num
    num = int(input("Enter a number (negative to stop): "))
    print("Total =", total)
```

```python
# The Password Loop
password = "1234"
guess = input("Enter password: ")
while guess != password:
    print("Wrong password, try again.") 
    guess = input("Enter password: ") 
    print("Access granted!")
```

```python
# Countdown
import time 
n = int(input("Enter starting number: ")) 
while n >= 0:
    print(n) 
    time.sleep(1) 
```

```python
 #Number of Digits
num = int(input("Enter an integer: ")) 
digits = 0
while num > 0:
    digits += 1 
    num //= 10
    print("Number of digits =", digits)
```

```python
#  Power without pow()
base = int(input("Enter base: "))
exp = int(input("Enter exponent: ")) 
result = 1
while exp > 0:
    result *= base
    exp -= 1
    print("Result =", result)
```

```python
# Binary to Decimal 
decimal = 0 
power = 0 
bit = int(input("Enter a bit (0 or 1, 2 to stop): "))
while bit != 2:
    decimal += bit * (2 ** power)
    power += 1
    bit = int(input("Enter a bit (0 or 1, 2 to stop): "))
    print("Decimal value =", decimal)
```

```python
#The Guessing Game
import random
secret = random.randint(1, 100) 
guess = int(input("Guess the number (1-100): "))
while guess != secret:
    if guess < secret:
        print("Higher!")
    else: print("Lower!")
    guess = int(input("Guess again: "))
print("Correct! The secret number was", secret)
```
