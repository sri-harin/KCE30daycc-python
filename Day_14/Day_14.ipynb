```python
#FULL PYRAMID
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

```python
#NUMBER PYRAMID
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i) + str(i) * (2 * i - 1))
```

```python
#INVERTED FULL PYRAMID
n = int(input("Enter number of rows: "))

for i in range(n, 0, -1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

```python
#HOLLOW SQUARE
n = int(input("Enter size: "))

for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
#DIAMOND PATTERN
n = int(input("Enter number of rows: "))

# Upper part
for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))

# Lower part
for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

```python
#X PATTERN
n = int(input("Enter size: "))

for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
#PASCAL'S TRIANGLE
n = int(input("Enter number of rows: "))

for i in range(n):
    num = 1
    for j in range(n - i):
        print(" ", end="")
    for j in range(i + 1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)
    print()
```

```python
#HOLLOW PYRAMID
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i), end="")
    for j in range(1, 2 * i):
        if j == 1 or j == 2 * i - 1 or i == n:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

```python
#HOURGLASS PATTERN
n = int(input("Enter number of rows: "))

# Upper inverted
for i in range(n, 0, -1):
    print(" " * (n - i) + "*" * (2 * i - 1))

# Lower
for i in range(2, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

```python
#PALINDROME NUMBER PYRAMID
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    print(" " * (n - i), end="")
    
    # Increasing
    for j in range(1, i + 1):
        print(j, end="")
        
    # Decreasing
    for j in range(i - 1, 0, -1):
        print(j, end="")
        
    print()
```

```python
#SUM OF DIGITS
num = int(input("Enter number: "))
sum_digits = 0

while num > 0:
    digit = num % 10
    sum_digits += digit
    num //= 10

print("Sum:", sum_digits)
```

```python
#PRODUCT OF DIGITS
num = int(input("Enter number: "))
product = 1

while num > 0:
    digit = num % 10
    product *= digit
    num //= 10

print("Product:", product)
```

```python
#REVERSE A NUMBER
num = int(input("Enter number: "))
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

print("Reverse:", reverse)
```

```python
#PALINDROME CHECK
num = int(input("Enter number: "))
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

if original == reverse:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#FIRST AND LAST DIGIT
num = int(input("Enter number: "))
last = num % 10

while num >= 10:
    num //= 10

first = num

print("First digit:", first)
print("Last digit:", last)
```

```python
#AMSTRONG NUMBER
num = int(input("Enter number: "))
original = num
sum_cube = 0

while num > 0:
    digit = num % 10
    sum_cube += digit ** 3
    num //= 10

if original == sum_cube:
    print("Armstrong Number")
else:
    print("Not Armstrong")
```

```python
#FREQUENCY DIGIT
num = int(input("Enter number: "))
digit_to_find = int(input("Enter digit to count: "))
count = 0

while num > 0:
    digit = num % 10
    if digit == digit_to_find:
        count += 1
    num //= 10

print("Frequency:", count)
```

```python
#STRONG NUMBER
import math

num = int(input("Enter number: "))
original = num
sum_fact = 0

while num > 0:
    digit = num % 10
    sum_fact += math.factorial(digit)
    num //= 10

if original == sum_fact:
    print("Strong Number")
else:
    print("Not Strong")
```

```python
#REPLACE ZEROS WITH ONES
num = int(input("Enter number: "))
result = 0
place = 1

while num > 0:
    digit = num % 10
    if digit == 0:
        digit = 1
    result += digit * place
    place *= 10
    num //= 10

print("Result:", result)
```
