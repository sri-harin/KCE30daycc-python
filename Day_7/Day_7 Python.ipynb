```python
#Alphabet Check
ch = input("Enter a character: ")

if ('a' <= ch <= 'z') or ('A' <= ch <= 'Z'):
    print("It is an alphabet")
else:
    print("It is NOT an alphabet")
```

```python
#Vowel or Consonant
ch = input("Enter a character: ").lower()

if 'a' <= ch <= 'z':
    if ch in ['a', 'e', 'i', 'o', 'u']:
        print("Vowel")
    else:
        print("Consonant")
else:
    print("Not an alphabet")
```

```python
#Digit or Not
ch = input("Enter a character: ")

if '0' <= ch <= '9':
    print("It is a digit")
else:
    print("It is NOT a digit")
```

```python
#Uppercase or Lowercase
ch = input("Enter a character: ")

if 'A' <= ch <= 'Z':
    print("Uppercase letter")
elif 'a' <= ch <= 'z':
    print("Lowercase letter")
else:
    print("Not an alphabet")
```

```python
#Special Character Hunt
ch = input("Enter a character: ")

if ('a' <= ch <= 'z') or ('A' <= ch <= 'Z') or ('0' <= ch <= '9'):
    print("Not a special character")
else:
    print("Special character")

```

```python
#ASCII Categorizer
ch = input("Enter a character: ")
ascii_val = ord(ch)

if ascii_val < 32:
    print("Control Character")
elif ascii_val <= 126:
    print("Printable Character")
else:
    print("Extended ASCII")
```

```python
#Case Converter
ch = input("Enter a character: ")

if 'a' <= ch <= 'z':
    print("Uppercase:", ch.upper())
elif 'A' <= ch <= 'Z':
    print("Lowercase:", ch.lower())
else:
    print("No conversion")
```

```python
#The “V” Challenge 
ch = input("Enter a character: ")

if ch == 'a' or ch == 'e' or ch == 'i' or ch == 'o' or ch == 'u' or \
   ch == 'A' or ch == 'E' or ch == 'I' or ch == 'O' or ch == 'U':
    print("Vowel")
else:
    print("Not a vowel")
```

```python
#Password Snippet
ch = input("Enter a character: ")

if ch != ' ' and (ch.isalpha() or ch.isdigit()):
    print("Safe character for password")
else:
    print("Not safe")
```

```python
#Alphabet Range
ch = input("Enter a character: ").upper()

if 'A' <= ch <= 'Z':
    if 'A' <= ch <= 'M':
        print("First half (A-M)")
    else:
        print("Second half (N-Z)")
else:
    print("Not an alphabet")
```

```python
#Day of the Week
day = int(input("Enter day number (1-7): "))

match day:
    case 1: print("Monday")
    case 2: print("Tuesday")
    case 3: print("Wednesday")
    case 4: print("Thursday")
    case 5: print("Friday")
    case 6: print("Saturday")
    case 7: print("Sunday")
    case _: print("Invalid input")

```

```python
#Month Name and days
month = int(input("Enter month number: "))

match month:
    case 1: print("January – 31 days")
    case 2: print("February – 28 days")
    case 3: print("March – 31 days")
    case 4: print("April – 30 days")
    case 5: print("May – 31 days")
    case 6: print("June – 30 days")
    case 7: print("July – 31 days")
    case 8: print("August – 31 days")
    case 9: print("September – 30 days")
    case 10: print("October – 31 days")
    case 11: print("November – 30 days")
    case 12: print("December – 31 days")
    case _: print("Invalid month")

```

```python
#Simple Calculator
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

match op:
    case '+': print("Result:", a + b)
    case '-': print("Result:", a - b)
    case '*': print("Result:", a * b)
    case '/': print("Result:", a / b if b != 0 else "Cannot divide by zero")
    case _: print("Invalid operator")
```

```python
#Vowel or Consonant
ch = input("Enter a character: ").lower()

match ch:
    case 'a' | 'e' | 'i' | 'o' | 'u':
        print("Vowel")
    case _:
        print("Consonant or Invalid")
```

```python
#Even or Odd
n = int(input("Enter a number: "))

match n % 2:
    case 0: print("Even")
    case 1: print("Odd")
```

```python
#Area Menu

choice = int(input("1.Circle 2.Square 3.Rectangle: "))

match choice:
    case 1:
        r = float(input("Enter radius: "))
        print("Area:", 3.14 * r * r)
    case 2:
        s = float(input("Enter side: "))
        print("Area:", s * s)
    case 3:
        l = float(input("Enter length: "))
        b = float(input("Enter breadth: "))
        print("Area:", l * b)
    case _:
        print("Invalid choice")
```

```python
#Gender Check
g = input("Enter gender (M/F): ")

match g:
    case 'M' | 'm': print("Male")
    case 'F' | 'f': print("Female")
    case _: print("Invalid input")
```

```python
#Default Case Demo
n = int(input("Enter number (1-3): "))

match n:
    case 1: print("One")
    case 2: print("Two")
    case 3: print("Three")
    case _: print("Invalid Input")
```

```python
#State of Water
temp = int(input("Enter temperature: "))

match temp // 100:
    case 0:
        if temp <= 0:
            print("Ice")
        else:
            print("Water")
    case 1:
        print("Steam")
    case _:
        print("Steam")
```
