```python
#CELCIUS TO FAHRENHEIT
c = float(input("Enter Celsius: "))
f = (c * 9/5) + 32
print("Fahrenheit:", f)
```

```python
#FAHRENHEIT TO CELCIUS
f = float(input("Enter Fahrenheit: "))
c = (f - 32) * 5/9
print("Celsius:", c)
```

```python
#DISTANCE CONVERTER
m = int(input("Enter meters: "))
km = m // 1000
rem = m % 1000
print(km, "km", rem, "m")
```

```python
#HEIGHT
feet = int(input("Enter feet: "))
inch = int(input("Enter inches: "))
total_inches = feet * 12 + inch
cm = total_inches * 2.54
print("Height in cm:", cm)
```

```python
#CURRENCY CONVERTER
usd = float(input("Enter USD: "))
rate = 83
inr = usd * rate
print("INR:", inr)
```

```python
#MINUTES TO HOUR AND MINUTES
mins = int(input("Enter minutes: "))
hours = mins // 60
rem = mins % 60
print(hours, "hours", rem, "minutes")
```

```python
#SECONDS TO HOUR,MINS,SECONDS
sec = int(input("Enter seconds: "))
hours = sec // 3600
sec %= 3600
minutes = sec // 60
seconds = sec % 60
print(hours, "H", minutes, "M", seconds, "S")

```

```python
#WEIHT
kg = float(input("Enter kg: "))
print("Grams:", kg * 1000)
print("Milligrams:", kg * 1000000)
```

```python
#STORAGE CONVERTER
gb = float(input("Enter GB: "))
mb = gb * 1024
kb = mb * 1024
byte = kb * 1024
print("MB:", mb)
print("KB:", kb)
print("Bytes:", byte)
```

```python
#SPEED
kmh = float(input("Enter speed (km/h): "))
ms = kmh * 5/18
print("m/s:", ms)
```

```python
#SWAP USING THIRD VARIABLE
a = int(input("Enter a: "))
b = int(input("Enter b: "))

temp = a
a = b
b = temp

print("a =", a)
print("b =", b)
```

```python
#SWAP WITHOUT THIRD VARIABLE
a = int(input("Enter a: "))
b = int(input("Enter b: "))

a = a + b
b = a - b
a = a - b

print(a, b)
```

```python
#SWAP USING MULTIPLICATION
a = int(input("Enter a: "))
b = int(input("Enter b: "))

a = a * b
b = a // b
a = a // b

print(a, b)
```

```python
#COMPOUND ADDITION
n = int(input("Enter number: "))
n += 10
print(n)
```

```python
#COMPOUND MULTIPLICATION
n = int(input("Enter number: "))
n *= 3
print(n)
```

```python
#MODULO ASSIGNMENT
a = int(input("Enter a: "))
b = int(input("Enter b: "))
a %= b
print(a)
```

```python
#TRIPLE SWAP
a = int(input("Enter a: "))
b = int(input("Enter b: "))
c = int(input("Enter c: "))

a, b, c = c, a, b
print(a, b, c)
```

```python
#XOR SWAP
a = int(input("Enter a: "))
b = int(input("Enter b: "))

a = a ^ b
b = a ^ b
a = a ^ b

print(a, b)
```
