```python
#FACTORIAL FINDER
n = int(input("Enter a number: "))
fact = 1

for i in range(1, n + 1):
    fact *= i

print("Factorial:", fact)
```

```python
#FIBONACCI SEQUENCE
n = int(input("Enter number of terms: "))

a, b = 0, 1

for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

```python
#SUM OF SQUARES
n = int(input("Enter N: "))
total = 0

for i in range(1, n + 1):
    total += i * i

print("Sum of squares:", total)
```

```python
#POWER SERIES
x = int(input("Enter x: "))
n = int(input("Enter n: "))
total = 0

for i in range(1, n + 1):
    total += x ** i

print("Sum:", total)
```

```python
#ALTERNATONG SERIES
n = int(input("Enter N: "))
total = 0

for i in range(1, n + 1):
    if i % 2 == 0:
        total -= i
    else:
        total += i

print("Result:", total)

```

```python
#HARMONIC SERIES
n = int(input("Enter N: "))
total = 0.0

for i in range(1, n + 1):
    total += 1 / i

print("Harmonic Sum:", total)
```

```python
#FACTORIAL SUM
n = int(input("Enter N: "))
fact = 1
total = 0

for i in range(1, n + 1):
    fact *= i
    total += fact

print("Factorial Sum:", total)

```

```python
#GEOMETRIC PROGRESSION
a = int(input("Enter first term: "))
r = int(input("Enter common ratio: "))
n = int(input("Enter number of terms: "))

term = a

for _ in range(n):
    print(term, end=" ")
    term *= r
```

```python
#SINE SERIES
import math

x = float(input("Enter x in radians: "))
n = int(input("Enter number of terms: "))

result = 0
sign = 1

for i in range(n):
    power = 2 * i + 1
    result += sign * (x ** power) / math.factorial(power)
    sign *= -1

print("sin(x) ≈", result)
```

```python
#ALL PRIME IN A RANGE
start = int(input("Start: "))
end = int(input("End: "))

for num in range(start, end + 1):
    if num > 1:
        is_prime = True
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            print(num, end=" ")
```

```python
#PERFECT NUMBERS
for num in range(1, 10001):
    total = 0
    for i in range(1, num):
        if num % i == 0:
            total += i
    if total == num:
        print(num)
```

```python
#PASCAL'S TRIALNGLE
n = int(input("Enter height: "))

for i in range(n):
    print(" " * (n - i), end="")
    num = 1
    for j in range(i + 1):
        print(num, end=" ")
        num = num * (i - j) // (j + 1)
    print()
```

```python
#NUM GUESSING GAME
import random

secret = random.randint(1, 100)
lives = 5

while lives > 0:
    guess = int(input("Guess the number: "))
    
    if guess == secret:
        print("You win!")
        break
    else:
        lives -= 1
        print("Wrong! Lives left:", lives)

if lives == 0:
    print("Game Over! Number was:", secret)
```

```python
#BINARY TO DECIMAL
binary = input("Enter binary number: ")
decimal = 0
power = 0

for digit in reversed(binary):
    decimal += int(digit) * (2 ** power)
    power += 1

print("Decimal:", decimal)
```

```python
#DECIMAL TO BINARY
n = int(input("Enter decimal number: "))
binary = ""

while n > 0:
    binary = str(n % 2) + binary
    n //= 2

print("Binary:", binary)
```

```python
#AMSTRONG NUMBER
for num in range(1, 1001):
    total = 0
    temp = num
    digits = len(str(num))
    
    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10
    
    if total == num:
        print(num)
```

```python
#SINE SERIES
import math

x = float(input("Enter x in radians: "))
term = x
sum_val = 0
n = 1
sign = 1

while abs(term) > 0.0001:
    sum_val += term
    n += 2
    sign *= -1
    term = sign * (x ** n) / math.factorial(n)

print("sin(x) ≈", sum_val)
```

```python
#NESTED MENU SYSTEM
while True:
    print("\n1. Math\n2. Patterns\n3. Exit")
    choice = int(input("Choose option: "))
    
    if choice == 1:
        print("1. Square\n2. Cube")
        sub = int(input("Choose: "))
        num = int(input("Enter number: "))
        
        if sub == 1:
            print("Square:", num ** 2)
        elif sub == 2:
            print("Cube:", num ** 3)
    
    elif choice == 2:
        rows = int(input("Enter rows: "))
        for i in range(1, rows + 1):
            print("*" * i)
    
    elif choice == 3:
        break
```
