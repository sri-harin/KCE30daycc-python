```python
#THE BASIC GRADE
percentage = float(input("Enter percentage: "))

if percentage > 90:
    print("Grade: A")
elif percentage > 80:
    print("Grade: B")
elif percentage > 70:
    print("Grade: C")
else:
    print("Grade: F")
```

```python
#ELECTRICITY BILL
units = int(input("Enter units consumed: "))
bill = 0

if units <= 100:
    bill = units * 0.50
elif units <= 300:
    bill = (100 * 0.50) + ((units - 100) * 0.75)
else:
    bill = (100 * 0.50) + (200 * 0.75) + ((units - 300) * 1.20)

print("Total Bill: ₹", bill)
```

```python
#INCOME TAX
income = float(input("Enter annual income: "))

if income < 250000:
    tax = 0
elif income <= 500000:
    tax = income * 0.05
else:
    tax = income * 0.20

print("Tax payable: ₹", tax)
```

```python
#BMI CATEGORIZER
weight = float(input("Enter weight (kg): "))
height = float(input("Enter height (m): "))

bmi = weight / (height ** 2)
print("BMI:", bmi)

if bmi < 18.5:
    print("Underweight")
elif bmi <= 24.9:
    print("Normal")
elif bmi <= 29.9:
    print("Overweight")
else:
    print("Obese")
```

```python
#TRIANGLES BY BANGLES
a = int(input("Enter angle 1: "))
b = int(input("Enter angle 2: "))
c = int(input("Enter angle 3: "))

if a > 0 and b > 0 and c > 0 and (a + b + c == 180):
    print("Valid Triangle")
else:
    print("Invalid Triangle")
```

```python
#TRIANGLE BY SIDES
a = int(input("Enter side a: "))
b = int(input("Enter side b: "))
c = int(input("Enter side c: "))

if a + b > c and a + c > b and b + c > a:
    print("Valid Triangle")
else:
    print("Invalid Triangle")
```

```python
#TRIANGLE TYPE
if a == b == c:
    print("Equilateral")
elif a == b or b == c or a == c:
    print("Isosceles")
else:
    print("Scalene")
```

```python
#RIGHT ANGLED TRIANGLE
sides = sorted([a, b, c])

if sides[0]**2 + sides[1]**2 == sides[2]**2:
    print("Right-angled Triangle")
else:
    print("Not Right-angled")
```

```python
#ACUTE/ABUTE/RIGHT
x, y, z = sides

if x**2 + y**2 == z**2:
    print("Right-angled")
elif x**2 + y**2 > z**2:
    print("Acute Triangle")
else:
    print("Obtuse Triangle")
```

```python
#POINT ON A CIRCLE
r = int(input("Enter radius: "))
x = int(input("Enter x: "))
y = int(input("Enter y: "))

value = x**2 + y**2

if value < r**2:
    print("Point is inside the circle")
elif value == r**2:
    print("Point is on the circle")
else:
    print("Point is outside the circle")
```

```python
#SQUARE OR TRIANGLE
length = int(input("Enter length: "))
breadth = int(input("Enter breadth: "))

if length == breadth:
    print("Square")
else:
    print("Rectangle")
```

```python
#COLLINEAR POINTS
x1, y1 = map(int, input("Enter x1 y1: ").split())
x2, y2 = map(int, input("Enter x2 y2: ").split())
x3, y3 = map(int, input("Enter x3 y3: ").split())

if (y2 - y1)*(x3 - x2) == (y3 - y2)*(x2 - x1):
    print("Collinear")
else:
    print("Not Collinear")
```

```python
#POINT RIGHT SIDE TRIANGLE
x1, y1 = map(int, input("Top-left (x y): ").split())
x2, y2 = map(int, input("Bottom-right (x y): ").split())
x, y = map(int, input("Point (x y): ").split())

if x1 <= x <= x2 and y2 <= y <= y1:
    print("Point is inside rectangle")
else:
    print("Outside rectangle")
```
