```python
#3*3 grid of *
for i in range(3):
    for j in range(3):
        print("*",end="")
    print()
```

```python
#COORDINATE MAX
for i in range(3):
    for j in range(3):
        print(f"{i},{j}",end="")
    print() 
```

```python
#N*N SQUARE
n=int(input("enter size:"))
ch=input("enter character:")

for i in range(n):
    for j in range(n):
        print(ch,end="")
        print()
```

```python
#MULTIPLICATION TABLE MATRIX
for i in range(1, 6):
    for j in range(1, 11):
        print(f"{i*j:3}", end=" ")
    print()
```

```python
#NUMBER SQUARE
for i in range(5):
    for j in range(1, 6):
        print(j, end=" ")
    print()
```

```python
#INCREMNTAL SQUARE
num = 1
for i in range(5):
    for j in range(5):
        print(num, end=" ")
        num += 1
    print()
```

```python
#RIGHT ANGLED STAR TRIANGLE
n = 5
for i in range(1, n+1):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
#INVENTED STAR TRIANGLE
n = 5
for i in range(n, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
#CHECKERED GRID
for i in range(5):
    for j in range(5):
        print((i + j) % 2, end=" ")
    print()
```

```python
#PRIME NUMBERS
for num in range(2, 101):
    prime = True
    for i in range(2, int(num**0.5)+1):
        if num % i == 0:
            prime = False
            break
    if prime:
        print(num, end=" ")
```

```python
#RIGHT TRIANGLE
n = 5
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
#RIGHT TRIANGLE
n = 5
for i in range(1, n+1):
    for j in range(i):
        print(i, end=" ")
    print()
```

```python
#INVERTED NUMBER TRIANGLE
n = 5
for i in range(n, 0, -1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
#FLOYDS TRIANGLE
n = 4
num = 1

for i in range(1, n+1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
#BINARY TRIANGLE
n = 5
for i in range(1, n+1):
    for j in range(1, i+1):
        print((i + j) % 2, end=" ")
    print()
```

```python
#ALPHABET TRIANGLE
n = 5
for i in range(1, n+1):
    for j in range(i):
        print(chr(65 + j), end=" ")
    print()
```

```python
#CONSTANT ALPHABET TRIANGLE
n = 5
for i in range(n):
    for j in range(i+1):
        print(chr(65 + i), end=" ")
    print()
```

```python
#RIGHT ANGLED STAR TRIANGLE
n = 5
for i in range(1, n+1):
    for space in range(n - i):
        print(" ", end=" ")
    for star in range(i):
        print("*", end=" ")
    print()
```

```python
#MIRRORED NUMBER TRIANGLE
n = 5
for i in range(n, 0, -1):
    for j in range(i):
        print(i, end=" ")
    print()
```
