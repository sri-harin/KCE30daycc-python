```python
#Print 1 to 10
i = 1
while True:
    print(i)
    i += 1
    if i > 10:
        break
```

```python
#Positive Number Only
while True:
    num = int(input("Enter a positive number: "))
    if num > 0:
        print("You entered:", num)
        break
```

```python
#The Range Lock
while True:
    num = int(input("Enter a number between 10 and 20: "))
    if 10 <= num <= 20:
        print("Valid input:", num)
        break

```

```python
#Add Until Zero
total = 0
while True:
    num = int(input("Enter a number (0 to stop): "))
    if num == 0:
        break
    total += num
print("Sum =", total)
```

```python
#Character Response
while True:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    print("Sum =", a + b)
    ch = input("Do you want to run again? (y/n): ")
    if ch.lower() != 'y':
        break
```

```python
#Password Guard
while True:
    pwd = input("Enter Password: ")
    if pwd == "secret":
        print("Access Granted!")
        break
```

```python
#Multiplication Table
while True:
    num = int(input("Enter a number: "))
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")
    ch = input("Do you want another table? (y/n): ")
    if ch.lower() != 'y':
        break
```

```python
#Digit Sum
num = int(input("Enter a number: "))
s = 0
while True:
    s += num % 10
    num //= 10
    if num == 0:
        break
print("Digit Sum =", s)
```

```python
#ATM Simulation
balance = 1000
while True:
    print("Balance:", balance)
    print("1. Withdraw\n2. Deposit")
    choice = int(input("Enter choice: "))
    if choice == 1:
        amt = int(input("Enter amount: "))
        balance -= amt
    elif choice == 2:
        amt = int(input("Enter amount: "))
        balance += amt
    cont = input("Continue transactions? (y/n): ")
    if cont.lower() != 'y':
        break
```

```python
#Early Exit
for i in range(1, 101):
    if i == 45:
        break
    print(i)
```

```python
#Skip the Unlucky Number
for i in range(1, 21):
    if i == 13:
        continue
    print(i)
```

```python
#Even Only
for i in range(1, 51):
    if i % 2 != 0:
        continue
    print(i)
```

```python
#The Searcher
arr = [2, 4, 7, 9, 11]
for num in arr:
    if num == 7:
        print("Found!")
        break
```

```python
#Stop on Negative
while True:
    num = int(input("Enter a number: "))
    if num < 0:
        break
```

```python
#Menu with Exit
while True:
    print("Menu:\n1. Option A\n2. Option B\n3. Exit")
    choice = int(input("Enter choice: "))
    if choice == 3:
        break
```

```python
#Prime Tester
N = int(input("Enter a number: "))
for i in range(2, N // 2 + 1):
    if N % i == 0:
        print("Not Prime")
        break
else:
    print("Prime")
```

```python
#Skip Multiples
for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        continue
    print(i)
```

```python

#Password with Attempts
for attempt in range(3):
    pwd = input("Enter Password: ")
    if pwd == "secret":
        print("Welcome")
        break
else:
    print("Account Locked")
```

```python
#Sum Until Limit
total = 0

while True:
    num = int(input("Enter a number: "))
    total += num
    if total > 100:
        break

print("Final Sum =", total)
```
