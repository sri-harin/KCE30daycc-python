```python
#REAL AND DISTINCT 
import math
a = int(input())
b = int(input())
c = int(input())

D = b*b - 4*a*c
print("Discriminant:", D)

root1 = (-b + math.sqrt(D)) / (2*a)
root2 = (-b - math.sqrt(D)) / (2*a)

print(root1, root2)
```

```python
#EQUAL ROOTS
import cmath  
a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))
D = b*b - 4*a*c

if D > 0:
    x1 = (-b + D**0.5) / (2*a)
    x2 = (-b - D**0.5) / (2*a)
    print("Distinct Real Roots =", x1, "and", x2)

elif D == 0:
    x = -b / (2*a)
    print("Equal Root =", x)

else:
    x1 = (-b + cmath.sqrt(D)) / (2*a)
    x2 = (-b - cmath.sqrt(D)) / (2*a)
    print("Complex Roots =", x1, "and", x2)
```

```python
# COMPLEX ROOTS
import math

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

D = b*b - 4*a*c

if D < 0:
    real = -b / (2*a)
    imag = math.sqrt(-D) / (2*a)
    print("Root 1 =", real, "+", imag, "i")
    print("Root 2 =", real, "-", imag, "i")
```

```python
# QUARDRATIC SOLVER
import math

a = float(input("Enter a: "))
b = float(input("Enter b: "))
c = float(input("Enter c: "))

if a == 0:
    print("Not a quadratic equation")
else:
    print(f"Equation: {a}x^2 + {b}x + {c} = 0")

    D = b*b - 4*a*c
    print("Discriminant =", D)

    if D > 0:
        x1 = (-b + math.sqrt(D)) / (2*a)
        x2 = (-b - math.sqrt(D)) / (2*a)
        print("Roots are real and distinct")
        print("Root 1 =", x1)
        print("Root 2 =", x2)

    elif D == 0:
        x = -b / (2*a)
        print("Roots are real and equal")
        print("Root =", x)

    else:
        real = -b / (2*a)
        imag = math.sqrt(-D) / (2*a)
        print("Roots are complex")
        print("Root 1 =", real, "+", imag, "i")
        print("Root 2 =", real, "-", imag, "i")
```
